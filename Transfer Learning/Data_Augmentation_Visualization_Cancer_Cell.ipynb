```
from keras.preprocessing import image
from keras.preprocessing.image import ImageDataGenerator
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```
img = image.load_img('cell.tif', target_size=(150, 150))
plt.imshow(img)
```

```
x = image.img_to_array(img)
```

```
x.shape
```

```
x = x.reshape((1,) + x.shape)
```

```
x.shape
```

## 1. Change input mean to 0 over the dataset

```
datagen1 = ImageDataGenerator(featurewise_center=True)

plt.imshow(image.array_to_img(datagen1.flow(x, batch_size=1)[0][0]))
```

```
datagen1.flow(x, batch_size=1)[0][0]
```

## 2. Set each sample mean to 0

```
datagen2 = ImageDataGenerator(samplewise_center=True)

plt.imshow(image.array_to_img(datagen2.flow(x, batch_size=1)[0][0]))
```

```
datagen2.flow(x, batch_size=1)[0][0]
```
