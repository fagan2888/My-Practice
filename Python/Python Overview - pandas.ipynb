# Reshape Data

### 1. melt
Turn columns into rows.

```python
data = pd.DataFrame([['Daniel', '-', 42], ['John', 12, 31], ['Jane', 24, 27]], columns=['name', 'treatment a', 'treatment b'])
data
```

**We can do either `df.melt()` or `pd.melt()`:**

```python
data.melt(id_vars='name', value_vars=['treatment a', 'treatment b'], var_name='treatment', value_name='result')
```

```python
pd.melt(data, id_vars='name', value_vars=['treatment a', 'treatment b'], var_name='treatment', value_name='result')
```

### 2. pivot
Turn unique values into separate columns. Index cannot contain duplicate entries.

```python
data = pd.DataFrame([['2010-01-30', 'tmax', 27.8], ['2010-01-30', 'tmin', 14.5], ['2010-02-02', 'tmax', 27.3], 
                     ['2010-02-02', 'tmin', 14.4]], columns=['date', 'element', 'value'])
data
```

**Cannot do `pd.pivot()`.**

```python
data.pivot(index='date', columns='element', values='value')
```

**Use `pivot_table` to apply aggregate functions.     
Both `pd.pivot_table()` and `df.pivot_table()` work:**

```python
pd.pivot_table(data, values='value', index='date', columns='element', aggfunc=np.mean)
```

```python
data.pivot_table(values='value', index='date', columns='element', aggfunc=np.mean)
```

**Another example for hierarchical indexes:**

```python
data = pd.DataFrame({"A": ["foo", "foo", "foo", "foo", "foo", "bar", "bar", "bar", "bar"],
                     "B": ["one", "one", "one", "two", "two", "one", "one", "two", "two"],
                     "C": ["small", "large", "large", "small", "small", "large", "small", "small", "large"],
                     "D": [1, 2, 2, 3, 3, 4, 5, 6, 7]})
data
```

```python
data.pivot_table(values='D', index=['A', 'B'], columns=['C'], aggfunc=np.sum)
```

### 4. merge
Combine horizontally. Like **JOIN** in SQL.

```python
data1 = pd.DataFrame({'state': ['California', 'Texas', 'Florida', 'New York'], 
                      'population': [39250017, 27862596, 20612436, 19745289]})
data1
```

```python
data2 = pd.DataFrame({'state': ['California', 'Florida', 'New York', 'Texas'], 
                      'ANSI': ['CA', 'FL', 'NY', 'TX']})
data2
```

```python
data3 = pd.DataFrame({'state': ['California', 'California', 'New York', 'New York'],
                      'city': ['San Diego', 'Sacramento', 'New York City', 'Albany']})
data3
```

```python
pd.merge(data1, data2, how='left', on='state')
```

```python
pd.merge(data3, data2, how='left', on='state')
```

```python
pd.merge(data3, data2, how='left', on='state', indicator='check')
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.datasets import load_iris
```

```python
data = pd.DataFrame(load_iris().data, columns=['Petal Length','Petal Width','Sepal Length','Sepal Width'])
data['Species'] = load_iris().target
data['Species'] = data['Species'].apply(lambda x: load_iris().target_names[x])
data.head()
```

```python
data.shape
```

### 1. plots

**scatter:**

```python
data.plot(kind='scatter', x='Sepal Length', y='Sepal Width')
plt.xlabel('sepal length (cm)')
plt.ylabel('sepal width (cm)')
```

**box:**

```python
data.plot(kind='box', y='Sepal Width')
plt.ylabel('sepal width (cm)')
```

```python
data.boxplot(column='Sepal Width', by='Species')
```

**histogram:**

```python
data.plot(kind='hist', y='Sepal Width', bins=50)
plt.xlabel('sepal length (cm)')
```

```python
data.plot(kind='hist', y='Sepal Width', bins=50, cumulative=True)
plt.xlabel('sepal length (cm)')
plt.title('Cumulative distribution function (CDF)')
```

```python
data[data.Species=='setosa'].plot(kind='hist', bins=100)
plt.title('Setosa data set')
data[data.Species=='versicolor'].plot(kind='hist', bins=100)
plt.title('Versicolor data set')
data[data.Species=='virginica'].plot(kind='hist', bins=100)
plt.title('Virginica data set')
```
